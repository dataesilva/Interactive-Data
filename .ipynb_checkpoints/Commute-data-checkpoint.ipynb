```python
import pandas as pd
```

```python
commute = pd.read_csv('commute.csv', header = 0)
```

```python
commute
```

```python
commute_clean = commute.astype({'Date': 'datetime64', 'Vehicle': 'category', 'Weather': 'category', 'Destination': 'category'})
```

```python
commute_clean['CommuteTime'].describe()
```

```python
pd.Series([x.year for x in commute_clean['Date']]).describe()
```

```python
commute_clean.groupby(['Vehicle', 'Weather', 'Destination']).describe()
```

```python
commute_clean.drop(labels = ['TripID'], axis = 1).groupby(['Vehicle', 'Weather']).describe()
```

```python
commute_clean.drop(labels = ['TripID'], axis = 1).groupby(['Vehicle']).describe()
```
