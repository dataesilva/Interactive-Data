```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
cces = pd.read_csv('cces_example_data.csv')
```

```python
cces = cces.astype({'sex': 'category', 'pid3': 'category', 'nonwhite': 'category', 'Evaluation': 'category'})
cces['Evaluation'].head()
```

```python
categories = ['Evaluation']
categories
```

```python
fig, ax = plt.subplots()
ax.bar([0,1,2,3], categories)

plt.show()
```

```python
heights = cces.groupby('Evaluation').count()['caseid']
```

```python
fig2, ax2 = plt.subplots()
ax2.bar([0,1,2,3], heights)
ax2.set_xticks([0, 1, 2, 3], ['one', 'two', 'three', 'four'])

plt.show()
```
